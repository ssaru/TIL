# Detection of time reversibility in time series by ordinal patterns analysis

<br/>

## Abstract

- real system에서 관찰되는 대부분의 time series data는 nonlinear를 상속받음
- 따라서 non-linearity 속성을 탐지하는 것은 대단히 관심분야가 될 수 밖에 없음
- non-linearity character를 확실히 해주는 대표적인 feature는 Time irreversibiliity 있음
- time series data가 reversible하다면, statistical properties가 시간 방향(direction of time)에 대해서 불변이다.
    - statistical properties
    Probability of any undetermined signal / PSD of a signal / ISI models / Ergodicity /Poison, Gaussian, Markov etc., processes / Auto correlation / Cross correlation / Moving average / Auto regressive

- 본 논문에서 서로 다른 스케일에서 선형 및 비선형 시계열의 시간 대칭성을 평가하기 위해 Ordinal Patterns (TiROP)에서 Time Reversibility를 제안한다.


## Introduction

- Time irreversibility는 non-equilibrium system과 non-conservative forces로부터 dynamics 결과에 대한 기본적인 속성이다.  


# 부가 내용

- 이러한 Time irreversibility는 파킨슨 환자의 떨리는 음성, 간질환자의 EEG데이터, 심장 조건에 따라 환자와 건강한 피험자로부터 추출 된 심박 간격 시간 시리즈 등에서 보고되었음
- 정적 인 선형 Gaussian 프로세스의 실현 인 모든 시계열은 공분산 함수의 대칭성 때문에 시간 가역성이 있습니다
- 그럼에도 불구하고, 비 가우스 진폭 분포는 고정 선형 가우스 프로세스의 정적 비선형 변환에 기인 할 수 있으며, 그 자체로는 시간적 비가역성의 증거가 될 수 없다.
- 또한 선형 시스템의 출력으로 모델링 된 비 가우스 프로세스는 가역적입니다.
- 대조적으로, 비 가우시안 잡음에 의해 여기 된 비선형 시스템의 출력은 시간 비가 역적이다
- 비선형 및 비 가우시안 선형 모델은 전형적으로 그들의 고차 의존성의 속성으로서 시간 방향성을 갖는다
- 따라서 시계열의 시간 가역성 특성에 대한 연구는 관측 된 데이터의 근본적인 비선형 메커니즘에 대한 의미있는 통찰력을 제공 할 수있다
- 고전적 시간 가역성 시험은 연구 된 신호 Xt의 고차 모멘트가 유한 일 필요가있다.
- 다른 테스트는 벡터 A와 시간 반전 된 버전 B의 분포를 직접 비교하거나 동적 인 투영을 유한 한 수의 평면에 비교하여 고안되었습니다
- 지난 몇 년 동안, $X_{t}$를 나타내는 1 차원 조경에서 점들 사이의 상호 가시성 관계와 같은 일부 작품은 소위 가시성 그래프를 기반으로 비가역성에 대한 통계적 테스트를 제안했습니다
- 이 연구들은 돌이킬 수없는 동역학이 그래프 속성의 확률 분포 사이의 비대칭을 초래한다는 것을 보여준다
- 최근에,이 접근법은 비 정적 프로세스의 연구를 위해 확장되었다
- 실제 가치의 시계열의 경우, Xt의 시간차 버전과 시간차 버전을 특성화하기 위해 다른 기호화 절차를 기반으로하는 시간 가역성 테스트가 제안 된 연구가 있습니다
- 이러한 심볼 변환은 일반적으로 시계열을 고유 한 패턴 또는 심볼의 개별 시퀀스로 변환하는 양자화 절차를 정의함으로써 수행됩니다
- 이러한 가역성 테스트 중 일부는 선험적 이항 통계를 사용하여 결과의 통계적 중요성을 평가합니다
- 그럼에도 불구하고, 그러한 테스트는 관측 된 심볼의 독립성을 가정하며, 이는 시간 상관과 함께 실제 데이터에서는 발생하지 않을 것입니다
- 이러한 일련의 상관 관계가있는 경우, 엄격한 이론적 틀을 도출 할 수 없으며, 시간 가역성 테스트의 유의 수준을 추정하기 위해 몬테카를로 시뮬레이션 (예 : 매개 변수 또는 비모수 재 리 샘플링)을 수행해야합니다.
- 이 연구에서 우리는 시간 순서의 전방 및 후방 통계의 경험적 분포를 비교하는 새로운 절차, Ordinal Patterns 방법으로부터의 Time Reversibility 방법을 제안한다
- 두 확률 분포 사이의 비대칭 성을 평가하기 위해 우리는 서수 기호 표현
- 기호 분석에 기반한 다른 접근 방식과는 달리 여기서 사용 된 서수 패턴 분석은 완전히 데이터 기반입니다. 즉 기호 변환에 선험적 임계 값이나 데이터 시퀀스에 대한 지식이 필요하지 않습니다.
- 우리는 기본 생성 프로세스에 대한 가정없이 대리 데이터 분석을 통해 시간 가역성 테스트를 완료합니다.
- 제안 된 프레임 워크는 선형, 비선형, 비 가우시안 확률 론적 및 결정 론적 프로세스로 시뮬레이션 된 합성 데이터에서 검증됩니다.

## Gaussian Process(GP)

- Supervised learning의 일종
- label이 된 data를 가지고 train시킨 후에, 새로운 data가 들어왔을 때, 그 data에 해당하는 label 혹은 output을 estimate하는 것
    - 이는 다른 supervised learning과 다르지 않음
<br/>

$X$ : input data  
$Y$ : output data  
$f$ : $X$를 $Y$ space로 mapping하는 함수  
$p$ : 확률 (posterior)
    
    
**Classification Problem with supervised learning**

$p(y|x, \theta) = f(y,x, \theta)$

<br/>

여기서 이 $f$를 찾는 것이 목적이고, 일반적으로 least square로 해당 문제를 풀 수 있다.  
  
$loss = (y - f(x, \theta))^2$

<br/>

### Gaussian Process

가우시안 프로세스는 $X$와 $Y$의 관계를 나타내는 어떠한 non-linear(혹은 linear)한 함수 $f$에 대한 distribution(분포)를 구한다.  
    
여기서 해당 함수의 공분산(covariance of $f$)는 uncertainty를 나타내며, 해당 함수 분포의 평균(mean)은 우리가 알고자하는 $\hat{Y}$는 우리가 알고자하는 추정값이 된다.
- 위에서 covariance가 uncertainty를 나타낸다는 이야기는 Gaussian이기 때문이다.

이를 수학적으로 표현하면,  
    
가우시안 프로세스는 함수에 대한 분포($p(f)$)를 정의한다. 여기서 $f$가 어떤 input space $\mathcal{X}$를 $\Re$로 mapping하는 함수다.

$f: \mathcal{X} \rightarrow \Re$

> $f$는 차원이 $\infty$일 수 있다.
    
$n$ 점($x_{i} \in \mathcal{X}$)에서 평가된 함수 값의 $n$차원 벡터를  
$f = (f(x_{1}), ..., f(x_{n}))$라고 하자
> 여기서 $f$는 *random variable*이다.

![gp](https://user-images.githubusercontent.com/13328380/56177621-39774c00-603a-11e9-970f-b3d7437f9cd7.PNG)
<br/>

---

Gaussian Process는 $X \Rightarrow Y$인 $f$의 분포(distribution)을 구한다.  
이 때, $N$개의 $(x, y)$ data set을 가지고 있다면 $p(f)$는 multivariate Gaussian distribution을 갖는다고 가정한다.  

#### Definition
$p(f)$는 어떤 유한한 부분집합(any finite subset) ${x_{1}, ... ,x_{n}} \subset \mathcal{X}$에 대해서 **Gaussian process**다.  
    
${x_{1}, ... ,x_{n}} \subset \mathcal{X}$일 때, 그 유한한 부분 집합 $p(f)$에 대한 marginal distribution은 multivariate Gaussian distribution을 갖는다.
    
---
    
여기서 $p(f)$가 multivariate Gaussian distribution을 갖는다고 가정했기 때문에 $p(f)$는 mean과 covariance로 나타낼 수 있다.

Gaussian Process에서 $p(f)$에서 $f$가 함수를 의미하기 때문에 여기서 **mean**과 **covariance** 또한 function이 되서,
**mean function**, **covariance function**이 된다. 

**mean function**은 구하기 쉽지만, **covariance function**은 구하기 어렵다.  
그래서 **covariance function**이 특정한 형태(parameter)를 갖고 있을 것이라고 가정하고, 이 형태의 함수를 **kernel** $K(x, x^{`})$이라고 부른다.    
    
<br/>

어떤 유한한 부분집합 ${x_{1},...,x_{n}\subset \mathcal{X}}$ 에서 $p(f)$는 Gaussian process이다.  
특정 유한한 부분집합 $p(f)$의 marginal distribution은 multivariate Gaussian distribution이다.  
    
Gaussian process (GPs)는 mean function $\mu(x)$와 커널이라고 불리는 covariance function $K(x, x^{'})$로 parameterize된다.  
     
$p(f(x), f(x^{'})) = N(\mu, \Sigma)$  
    
$\mu = \begin{bmatrix} \mu(x) \\ \mu(x^{'}) \end{bmatrix}$  
$\Sigma = \begin{bmatrix} K(x, x) & K(x, x^{'}) \\ K(x^{'}, x) & K(x^{'}, x^{'}) \end{bmatrix}$

#### random variable
수학적으로 확률 변수는 표본 공간을 정의역(domain)으로 가지고 실수를 공역(range)으로 갖는 함수로 정의한다.  
즉 확률 변수는 표본 공간의 모든 표본에 대해 어떤 실수 값을 붙인(할당한)것이다.  

$$\omega \in \Omega \xrightarrow{X} x \in R$$  
     
$$X(\omega) = x (x \in R)$$

> $X, Y$등의 대문자 알파벳을 사용하여 확률 변수를 표기하고 확률 변수에 의해 할당된 실수는 $x, y$와 같이 소문자 알파벳으로 표기한다.
> 경우에 따라서는 소문자 알파벳으로 확률 변수를 표기할 수도 있다.  
    
#### Marginal Probability

한글표기로는 주변확률분포. 어떤 변수들에 대한 확률분포(probability distribution)을 알고 있을 때, 그 변수들의 하위 집합(subset)의 분포를 알아야할 때가 있다.  
이러한 하위 집합(subset)에 대한 확률 분포를 *marginal probability distribution*이라고 한다.  
    
> 예를들어 이산형 변수 $x, y$가 있고, $P(x, y)$를 알고 있을 때, 덧셈 규칙을 이용해서 P(x)를 구할 수 있다.  
    
> $\forall{x}, P(x=x) = \sum_{y}P(x=x, y=y)$  

> $p(x)=\int{P_{x, y}}dy$          

#### Multivariate normal distribution
다변수 정규분포, 다변량 가우스 분포 또는 joint normal distribution과 같은 1차원(univariate; 단변수)정규 분포를 더 높은 차원으로 일반화한 것

$k$-dimensional random vector $X = (X_{1},...,X_{k})^T$의 multivariate normal distribution은 다음과 같은 notation을 따른다.

$X \sim \mathcal{N}(\mu, \Sigma)$  
     
만약 $X$의 $k$-dimension을 알고있다면, 다음과 같이 표기할 수 있다.
     
$X \sim \mathcal{N}_{k}(\mu, \Sigma)$   
    
여기서 $k$-dimensioanl mean vector는 다음과 같다.  
    
$\mu = E[X] = [E[X_{1}], E[X_{2}], ..., E[X_{k}]]^T$  
    
$kxk$ covariance matrix는 아래와 같이 구할 수 있다.  
$\Sigma_{i, j} =: E[(X_{i} - \mu_{i})(X_{j} - \mu_{j})] = Cov[X_{i}, X_{j}]$
$\Sigma =: E[(X - \mu)(X - \mu)]^{T} = Cov[X_{i}, X_{j}; 1 \leq i, j \leq k]$
