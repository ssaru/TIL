# Detection of time reversibility in time series by ordinal patterns analysis

<br/>

## Abstract

- real system에서 관찰되는 대부분의 time series data는 nonlinear를 상속받음
- 따라서 non-linearity 속성을 탐지하는 것은 대단히 관심분야가 될 수 밖에 없음
- non-linearity character를 확실히 해주는 대표적인 feature는 Time irreversibiliity 있음
- time series data가 reversible하다면, statistical properties가 시간 방향(direction of time)에 대해서 불변이다.
    - statistical properties
    Probability of any undetermined signal / PSD of a signal / ISI models / Ergodicity /Poison, Gaussian, Markov etc., processes / Auto correlation / Cross correlation / Moving average / Auto regressive

- 본 논문에서 서로 다른 스케일에서 선형 및 비선형 시계열의 시간 대칭성을 평가하기 위해 Ordinal Patterns (TiROP)에서 Time Reversibility를 제안한다.

### Time reversibility vs Time irreversibility
Time reversibility와 irreversibility는 임의의 time series data $x(t),\  t ∈ [0, T]$가 있을 때,  $x^{t.r.}(t) = x(T − t)$를 만족하면 Time reversibility라고 하며, 만족하지 않으면 irreversibility라고 한다.

### Dynamics resulting from non-conservative forces
물리학에서 에너지 보존의 법칙이 있는데, 실세계에서 보존된 에너지들을 보존되지 않게 만드는 힘들(마찰력 등등)

<img src="https://user-images.githubusercontent.com/13328380/56091743-797ce880-5eed-11e9-9226-897782cc9b17.jpg" width="500" height="500">

- 이러한 Time irreversibility는 파킨슨 환자의 떨리는 음성, 간질환자의 EEG데이터, 심장 조건에 따라 환자와 건강한 피험자로부터 추출 된 심박 간격 시간 시리즈 등에서 보고되었음
- 정적 인 선형 Gaussian 프로세스의 실현 인 모든 시계열은 공분산 함수의 대칭성 때문에 시간 가역성이 있습니다
- 그럼에도 불구하고, 비 가우스 진폭 분포는 고정 선형 가우스 프로세스의 정적 비선형 변환에 기인 할 수 있으며, 그 자체로는 시간적 비가역성의 증거가 될 수 없다.
- 또한 선형 시스템의 출력으로 모델링 된 비 가우스 프로세스는 가역적입니다.
- 대조적으로, 비 가우시안 잡음에 의해 여기 된 비선형 시스템의 출력은 시간 비가 역적이다
- 비선형 및 비 가우시안 선형 모델은 전형적으로 그들의 고차 의존성의 속성으로서 시간 방향성을 갖는다
- 따라서 시계열의 시간 가역성 특성에 대한 연구는 관측 된 데이터의 근본적인 비선형 메커니즘에 대한 의미있는 통찰력을 제공 할 수있다
- 고전적 시간 가역성 시험은 연구 된 신호 Xt의 고차 모멘트가 유한 일 필요가있다.
- 다른 테스트는 벡터 A와 시간 반전 된 버전 B의 분포를 직접 비교하거나 동적 인 투영을 유한 한 수의 평면에 비교하여 고안되었습니다
- 지난 몇 년 동안, $X_{t}$를 나타내는 1 차원 조경에서 점들 사이의 상호 가시성 관계와 같은 일부 작품은 소위 가시성 그래프를 기반으로 비가역성에 대한 통계적 테스트를 제안했습니다
- 이 연구들은 돌이킬 수없는 동역학이 그래프 속성의 확률 분포 사이의 비대칭을 초래한다는 것을 보여준다
- 최근에,이 접근법은 비 정적 프로세스의 연구를 위해 확장되었다
- 실제 가치의 시계열의 경우, Xt의 시간차 버전과 시간차 버전을 특성화하기 위해 다른 기호화 절차를 기반으로하는 시간 가역성 테스트가 제안 된 연구가 있습니다
- 이러한 심볼 변환은 일반적으로 시계열을 고유 한 패턴 또는 심볼의 개별 시퀀스로 변환하는 양자화 절차를 정의함으로써 수행됩니다
- 이러한 가역성 테스트 중 일부는 선험적 이항 통계를 사용하여 결과의 통계적 중요성을 평가합니다
- 그럼에도 불구하고, 그러한 테스트는 관측 된 심볼의 독립성을 가정하며, 이는 시간 상관과 함께 실제 데이터에서는 발생하지 않을 것입니다
- 이러한 일련의 상관 관계가있는 경우, 엄격한 이론적 틀을 도출 할 수 없으며, 시간 가역성 테스트의 유의 수준을 추정하기 위해 몬테카를로 시뮬레이션 (예 : 매개 변수 또는 비모수 재 리 샘플링)을 수행해야합니다.
- 이 연구에서 우리는 시간 순서의 전방 및 후방 통계의 경험적 분포를 비교하는 새로운 절차, Ordinal Patterns 방법으로부터의 Time Reversibility 방법을 제안한다
- 두 확률 분포 사이의 비대칭 성을 평가하기 위해 우리는 서수 기호 표현
- 기호 분석에 기반한 다른 접근 방식과는 달리 여기서 사용 된 서수 패턴 분석은 완전히 데이터 기반입니다. 즉 기호 변환에 선험적 임계 값이나 데이터 시퀀스에 대한 지식이 필요하지 않습니다.
- 우리는 기본 생성 프로세스에 대한 가정없이 대리 데이터 분석을 통해 시간 가역성 테스트를 완료합니다.
- 제안 된 프레임 워크는 선형, 비선형, 비 가우시안 확률 론적 및 결정 론적 프로세스로 시뮬레이션 된 합성 데이터에서 검증됩니다.
- 이 방법은 다른 실시간 시리즈의 모음에도 설명되어 있습니다. 우리의 방법의 신뢰성과 성능은 고전적인 순간 기반 방법에 의해 얻어진 것과 비교된다.
- 논문의 나머지 부분은 다음과 같이 구성된다. 섹션 II는 우리의 솔루션을 벤치마킹하기 위해 사용 된 비교 방법과 제안 된 프레임 워크를 설명한다. 실험적 결과와 합성 시계열에서의 방법 평가는 III 절에있다. 실제 데이터에 대한 테스트의 평가는 섹션 IV에서 제공됩니다. 마지막으로, 우리는 섹션 V에서 논문에 대한 결론과 디스커션을 진행한다.
